```python
def format_date(date):
    try:
        formatted_date = pd.to_datetime(date, format='%d %b %Y')
    except:
        month_names = {'janvier':'Jan', 'février':'Feb', 'fév':'Feb', 'fev':'Feb', 'mars':'Mar', 'avril':'Apr', 'avr':'Apr', 'mai':'May', 'juin':'Jun', 'juillet':'Jul', 'juil':'Jul', 'aout':'Aug', 'août':'aout', 'septembre':'Sep', 'sept':'Sep', 'octobre':'Oct', 'novembre':'Nov', 'décembre':'Dec', 'déc':'Dec'}
        day = date.split()[0]
        month = date.split()[1].lower()
        year = date.split()[2]
        if month not in month_names.values() and month not in [ele.lower() for ele in list(month_names.values())]:
            month = month_names[month]
        formatted_date = pd.to_datetime(' '.join([day, month, year]), format='%d %b %Y')
    return formatted_date
```
